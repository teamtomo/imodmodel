```python
from imodmodel import ImodModel
from imodmodel.models import Mesh, Object
import trimesh
```

```python
teapot = trimesh.load('teapot.obj')
```

```python
model = ImodModel()
model.objects.append(Object())
model.objects[0].meshes.append(Mesh())
model.objects[0].meshes[0].vertices = teapot.vertices
model.objects[0].meshes[0].indices = teapot.faces
```

```python
model.to_file('teapot.mod')
```
